# Data Cleaning

```python
df = dt.copy()
```

```python
#Converting the rows of visual acuity measuremnt in fraction(snellen) format to log(LogMAR) format
df['Visual Acuity Measurements'] = df['Visual Acuity Measurements'].replace(['20/40', '20/20'],
                                                                           ['LogMAR 0.3', 'LogMAR 0.0'])
df['Visual Acuity Measurements'].unique()
```

```python
#Replacing 'glaucoma in the family' with 'glaucoma'
df['Medical History'] = df['Medical History'].replace('Glaucoma in family', 'Glaucoma')
df['Medical History'].unique()
```

```python
#unpacking the Medication Usage column
medList = []

for val in df['Medication Usage']:
    if pd.notna(val):  # Check if the value is not NaN
        if isinstance(val, str):  # Check if the value is a string
            if ',' in val:
                mList = val.split(',')
                meds = [x.strip() for x in mList]
                medList.extend(meds)
            else:
                val.strip()
                medList.append(val)
        else:
            # Handle cases where the value is not a string (e.g., float)
            # You might want to add additional handling based on your specific case
            continue

medList
```

```python
medList_copy = list(set(medList))
medList_copy
```

```python
medPos = []

for positions in df['Medication Usage'].values:
    medTable = [0,0,0,0,0,0,0]
    if pd.notna(positions):
        for p in medList_copy:
            if p in positions:
                Pidx = medList_copy.index(p)
                medTable[Pidx] = 1
        medPos.append(medTable)
    else:
        medPos.append(medTable)
medPos
```

```python
dfmed = pd.DataFrame(medPos, columns = medList_copy) # converting the postions list to a dataframe
dfmed
```

```python
df = df.join(dfmed)
```

```python
#unpacking the Visual Field Test Result column
senList = []

for val in df['Visual Field Test Results']:
    if ',' in val:
        visList = val.split(',')
        sensitivity = visList[0].split(':')
        senList.append(sensitivity[-1].strip())
senList
```

```python
#unpacking the Visual Field Test Result column
speList = []

for val in df['Visual Field Test Results']:
    if ',' in val:
        visList = val.split(',')
        specificity = visList[1].split(':')
        speList.append(specificity[-1].strip())
speList
```

```python
df.insert(10, 'sensitivity', senList)
df[['Visual Field Test Results', 'sensitivity']]
```

```python
df.insert(11, 'specificity', speList)
df[['Visual Field Test Results', 'specificity']]
```
