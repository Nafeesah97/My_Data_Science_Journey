```python
#Data collection
def dataCollector():
    '''
    This function accepts user first value, second value and what type of
    operation to be carried out. It returns the data entered by the user.
    '''
    data = []
    userVal1 = input('Type first value    ')
    userVal2 = input('Type second value    ')
    operator = input('choose operator    ')
    data.append(userVal1)
    data.append(userVal2)
    data.append(operator)
    return data


#Datacheck
def inputChecker(value):
    '''
    This function checks the user's value to be sure
    the operation can be carried out on it by converting it to a float character.
    It returns the value converted to a float character.
    '''
    try:
        # Convert it into float
        check = float(value)
        return check
    except ValueError:
        print(f'{value} is not a number. Please input a number!')

#Operator
def operation(val1, val2, operator):
    '''
    This function accepts the data entered by the user and perform the
    calculation. It returns the result of the operation on the 2 values
    entered by the user.
    '''
    while True:
        try:
            if operator == '*':
                res = float(val1) * float(val2)
            elif operator == '/':
                res = float(val1) / float(val2)
            elif operator == '-':
                res = float(val1) - float(val2)
            elif operator == '+':
                res = float(val1) + float(val2)
            elif operator == '%':
                res = float(val1) % float(val2)
            else:
                res = 'Operation cannot be performed'
            return res
        except TypeError:
            pass
            res = 'Error'
            return res

#To quit or continue calculation
def calcAgain():
    '''
    This function asks the user if they want to continue calculating
    or will quit.
    '''
    status = input('Do you wish to continue? Type Y for yes or any other key to quit    ')
    if status.upper() == 'Y':
        mainFunction()
    else:
        pass
        print('Thanks for using')

#Main calculation
def mainFunction():
    '''
    This function collects user's inputs, check if values and operation
    are valid and perform operation. It displays result of operation
    '''
    num = dataCollector()
    operator = num[2]
    val1 = inputChecker(num[0])
    val2 = inputChecker(num[1])
    finalResult = operation(val1, val2, operator)
    print(f'Result = {finalResult}')
    calcAgain()

#Calculator generator
def calculator(result = None):
    '''
    This function accepts  user's name. It displays info about the
    calculator and result.
    '''
    userName = input('what is your name?   ')
    print(f'''
    Welcome {userName}!, this calculator does the following operations:
    multiplication(*), division(/), subtraction(-), addition(+) and remainder(%).
    Input 2 values and indicate what type of operation by inputting their signs
    ''')
    mainFunction()
    return result


calculator(result = None)
```
